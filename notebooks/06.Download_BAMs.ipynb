# Download all BAMs

`pysradb download` will download the `.sra` files by default. To specify an alternate protocol specify the column name via `--col`. For example the `sra_url_alt1` above contains the BAM locations which is what we want to download:

```
!pysradb metadata SRP238587 --detailed | pysradb download --col sra_url_alt1
```
